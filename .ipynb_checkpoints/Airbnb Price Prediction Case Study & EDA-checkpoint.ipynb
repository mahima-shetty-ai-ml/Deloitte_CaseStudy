# Airbnb Price Case Study Exploratory Data Analysis

Airbnb is an online marketplace that connects people who want to rent out their homes with people who are looking for accommodations in that locale. Airbnb is the platform where this happens. It's part “sharing economy,” part entrepreneurship, part meeting people, and sharing experiences.
The challenge statement was AirBNB price prediction.Some of the factors were property, room type, amenities, accommodates, bathroom, bed_type, etc.
Using machine learning technique, proceeding with all factors leading to the price prediction to develop the model.


```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
train_data = pd.read_csv('train.csv')
test_data = pd.read_excel('test.xlsx')
```

## Dropping unnecessary columns out from dataset

```python
train_data = train_data.drop([' '], axis=1)
test_data = test_data.drop([' '], axis = 1)
```

## Evaluating Columns and its purpose

```python
# print(train_data.head())
# print(test_data.head())

print(train_data.columns)
print(test_data.columns)
print(train_data.shape)
print(test_data.shape)
```

```python
test_data['log_price'] = np.nan
```

## Converting "host_since" column into proper datetime format, filling up na values with forward fill and extracting only year for quick analysis  

```python
train_data['host_since'] = pd.to_datetime(train_data['host_since'],format='%m/%d/%Y')
train_data["host_year"] = train_data['host_since'].dt.year

test_data['host_since'] = pd.to_datetime(test_data['host_since'],format='%Y-%m-%d %H:%M:%S')
test_data['host_year'] = test_data['host_since'].dt.year
```

```python
train_data['host_year']=train_data['host_year'].fillna(method="ffill")
test_data['host_year']=test_data['host_year'].fillna(method="ffill")
```

```python
train_data['host_year'] = train_data['host_year'].apply(np.int64)
test_data['host_year'] = test_data['host_year'].apply(np.int64)
```

## Insights 1: Which host year's property had the good economy log_prices?




```python
train_data['host_year'].unique()
```

```python
list1 = []
for i in [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]:
    val = train_data[train_data['host_year']==i]['log_price'].sum()
    list1.append(val)
print(list1)
```

```python
fig = plt.figure()
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]

ax = fig.add_axes([0,0,1,1])
# ax.bar(years,list1)
ax.bar(years, list1, color = 'green')
plt.title("Log Price Sum total wrt Host_year")
plt.show()
```

Prices were dependent upon the few important aspects in the dataset which will be explained later in this code.

The host year 2015 has the highest log_prices followed by the year 2014 and the year 2016. The least mean log_price value can be seen for the host year 2011 and the host year 2017.

# Insights 4: Which room_type has the good Price for occupancy?
